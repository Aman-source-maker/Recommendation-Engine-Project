# MovieLens Movie Recommendation Engine`n`nThis notebook uses the MovieLens dataset from `Desktop/Movielens/ml-latest-small` to recommend movies based on movies a user has already watched and liked.`n`nThe approach uses item-item collaborative filtering: movies are compared by rating patterns across users, then recommendations are generated from similar movies.

```python
import os`nfrom pathlib import Path`n`nimport numpy as np`nimport pandas as pd`n`npd.set_option("display.max_colwidth", 120)`n`ndesktop = Path.home() / "OneDrive" / "Desktop"`nif not desktop.exists():`n    desktop = Path.home() / "Desktop"`n`ndata_dir = desktop / "Movielens" / "ml-latest-small"`nratings_path = data_dir / "ratings.csv"`nmovies_path = data_dir / "movies.csv"`n`nprint("Using dataset folder:", data_dir)`nprint("Ratings exists:", ratings_path.exists())`nprint("Movies exists:", movies_path.exists())
```

```python
ratings = pd.read_csv(ratings_path)`nmovies = pd.read_csv(movies_path)`n`nprint("Ratings shape:", ratings.shape)`nprint("Movies shape:", movies.shape)`ndisplay(ratings.head())`ndisplay(movies.head())
```

```python
# Build a user-movie rating matrix.`n# Rows are users, columns are movies, values are ratings.`nuser_movie = ratings.pivot_table(index="userId", columns="movieId", values="rating")`n`n# Keep popularity stats so weakly-rated movies do not dominate recommendations.`nmovie_stats = ratings.groupby("movieId").agg(`n    rating_count=("rating", "count"),`n    rating_mean=("rating", "mean")`n).reset_index()`n`nmovie_lookup = movies.merge(movie_stats, on="movieId", how="left").fillna({"rating_count": 0, "rating_mean": 0})`n`nprint("User-movie matrix shape:", user_movie.shape)`ndisplay(movie_lookup.sort_values("rating_count", ascending=False).head(10))
```

```python
def search_movies(query, limit=10):`n    """Search movies by partial title."""`n    query = query.lower().strip()`n    result = movie_lookup[movie_lookup["title"].str.lower().str.contains(query, regex=False, na=False)]`n    cols = ["movieId", "title", "genres", "rating_count", "rating_mean"]`n    return result.sort_values(["rating_count", "rating_mean"], ascending=False)[cols].head(limit)`n`n# Example search`nsearch_movies("toy story")
```

```python
def recommend_from_movies(watched_movie_ids, top_n=10, min_common_users=20, min_rating_count=25):`n    """`n    Recommend movies similar to the movies in watched_movie_ids.`n`n    watched_movie_ids: list of MovieLens movieId values the user watched and liked.`n    top_n: number of recommendations to return.`n    min_common_users: minimum overlapping users required for similarity.`n    min_rating_count: minimum total ratings for recommended movies.`n    """`n    watched_movie_ids = [mid for mid in watched_movie_ids if mid in user_movie.columns]`n    if not watched_movie_ids:`n        raise ValueError("None of the watched_movie_ids were found in the rating matrix.")`n`n    seed_profile = user_movie[watched_movie_ids].mean(axis=1)`n    scores = []`n`n    for movie_id in user_movie.columns:`n        if movie_id in watched_movie_ids:`n            continue`n`n        pair = pd.concat([seed_profile, user_movie[movie_id]], axis=1).dropna()`n        common_users = len(pair)`n        if common_users < min_common_users:`n            continue`n`n        similarity = pair.iloc[:, 0].corr(pair.iloc[:, 1])`n        if pd.isna(similarity):`n            continue`n`n        scores.append((movie_id, similarity, common_users))`n`n    recs = pd.DataFrame(scores, columns=["movieId", "similarity", "common_users"])`n    recs = recs.merge(movie_lookup, on="movieId", how="left")`n    recs = recs[recs["rating_count"] >= min_rating_count]`n`n    # Blend similarity with general quality and popularity for more stable recommendations.`n    recs["score"] = (`n        recs["similarity"] * 0.70 +`n        (recs["rating_mean"] / 5.0) * 0.20 +`n        np.log1p(recs["rating_count"]) / np.log1p(recs["rating_count"].max()) * 0.10`n    )`n`n    cols = ["movieId", "title", "genres", "score", "similarity", "common_users", "rating_count", "rating_mean"]`n    return recs.sort_values("score", ascending=False)[cols].head(top_n).reset_index(drop=True)`n
```

```python
def recommend_by_titles(title_queries, top_n=10):`n    """Search each title query, pick the most popular match, and recommend from those movies."""`n    matched_ids = []`n    matches = []`n`n    for query in title_queries:`n        result = search_movies(query, limit=1)`n        if result.empty:`n            print(f"No match found for: {query}")`n            continue`n        row = result.iloc[0]`n        matched_ids.append(int(row["movieId"]))`n        matches.append(row)`n`n    if not matched_ids:`n        raise ValueError("No movie titles matched.")`n`n    print("Matched input movies:")`n    display(pd.DataFrame(matches)[["movieId", "title", "genres", "rating_count", "rating_mean"]])`n    return recommend_from_movies(matched_ids, top_n=top_n)`n`n# Easier title-based example. Edit the list below.`nrecommend_by_titles(["matrix", "dark knight", "inception"], top_n=10)
```
